```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

load dataset


```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

checking outliers

```python
# Checking for Outliers
# Outliers can significantly impact the performance of machine learning models. We can use box plots to visualize potential outliers:
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize outliers using box plots for all features
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Box Plot of Brest Cancer Dataset Features")
plt.show()
```

removing outliers


```python

Q1 = df['mean radius'].quantile(0.25)
Q3 = df['mean radius'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_cleaned = df[(df['mean radius'] >= lower_bound) & (df['mean radius'] <= upper_bound)]
outlier = df[(df['mean radius'] < lower_bound) & (df['mean radius'] > upper_bound)]
outlier
```

```python
plt.figure(figsize=(15, 10))
for i, column in enumerate(df_cleaned.columns[:-1]):
    plt.subplot(5, 6, i + 1)
    sns.histplot(df_cleaned[column], kde=True)
    plt.title(column)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import boxcox

# Apply Box-Cox transformation to positively skewed features (example: 'mean area')
df_cleaned['mean area'], _ = boxcox(df_cleaned['mean area'] + 1)  # Adding 1 to avoid log(0)
```

scaling

```python

from sklearn.preprocessing import StandardScaler

X = df_cleaned.drop('target', axis=1)
y = df_cleaned['target']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

classification algorithm implimentation


slitting the dataset for training and testing

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(X_train, y_train)
```

k-Nearest Neighbors (k-NN)

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
```

Model comparison

```python

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "k-NN": KNeighborsClassifier()
}

def evaluate_model(model):

    model.fit(X_train, y_train)

    predictions = model.predict(X_test)

    accuracy = accuracy_score(y_test, predictions)

    report = classification_report(y_test, predictions, target_names=["Malignant", "Benign"])

    conf_matrix = confusion_matrix(y_test, predictions)

    return accuracy, report, conf_matrix


results = {}

for name, model in models.items():
    accuracy, report, conf_matrix = evaluate_model(model)
    results[name] = {
        "Accuracy": accuracy,
        "Report": report,
        "Confusion Matrix": conf_matrix
    }
```

visualize the models

```python
import matplotlib.pyplot as plt

for name, metrics in results.items():
    print(f"Model: {name}")
    print(f"Accuracy: {metrics['Accuracy']:.4f}")
    print("Classification Report:\n", metrics['Report'])
    print("Confusion Matrix:\n", metrics['Confusion Matrix'])
    sns.heatmap(conf_matrix, annot=True)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix for the Breast Cancer Dataset')
    plt.show()
    print("=" * 50)
```
